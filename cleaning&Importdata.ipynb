```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
df=pd.read_csv("walmart_data.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Date"]=pd.to_datetime(df["Date"],dayfirst=True).dt.date
```

```python
df["Time"]=pd.to_datetime(df["Time"],format="%H:%M:%S").dt.time
```

```python
df.rename(columns={"Date":"Sale_date","Time":"Sale_time"})
```

```python
df.duplicated().sum()
```

```python
df.describe()
```

```python
mysql_engine=create_engine('mysql+pymysql://root@localhost:3306/walmart_sales')
```

```python
df.to_sql(name="walmart_data",con=mysql_engine,if_exists="replace",index=False)
```
